```python
import os

from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent

from digikala_tools.langchain_tools import get_digikala_search_tool

load_dotenv()
```

```python
model = ChatOpenAI(
    model=os.getenv("OPENAI_LLM_MODEL"),
    api_key=os.getenv("OPENAI_API_KEY"),
)

digikala_search_tool = get_digikala_search_tool(return_direct=True)
agent = create_react_agent(
    model=model,
    tools=[digikala_search_tool],
    prompt="You are a helpful assistant",
)

prompt = "لیست گوشی های بین ۱۰ تا ۲۰ میلیون تومن رو برام پیدا کن"

result = agent.stream(
    {
        "messages": [
            {
                "role": "user",
                "content": prompt,
            }
        ]
    },
    stream_mode="values",
)
for r in result:
    msg = r["messages"][-1]
    msg.pretty_print()
```
